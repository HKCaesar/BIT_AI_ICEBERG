```python
import numpy as np

from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import *
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential
from keras.models import Model
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense, ZeroPadding2D
from keras.layers import BatchNormalization
from keras.layers import concatenate
from keras.layers import Input
from keras import initializers
from keras import regularizers
from keras.layers.advanced_activations import LeakyReLU
from sklearn.model_selection import StratifiedKFold
from keras.utils.np_utils import to_categorical
import json
import pandas as pd
import random
from keras import backend as K
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
K.set_image_dim_ordering('tf')

data = pd.read_json('train.json')

data = data.drop('id', 1)
data = data.drop('inc_angle', 1)

```

```python
Y = data['is_iceberg']

data = data.drop('is_iceberg', 1)
```

```python
X1 = data['band_1']
X2 = data['band_2']
X1 = np.array(X1)
X2 = np.array(X2)
```

```python
# X = []

# for i in range(len(X1)):
#     X.extend(X1[i])
    
# X = np.array(X)
# X = np.reshape(X, (len(X1), 75, 75, 1))

# X.shape
```

```python
X_R1 = np.array([np.array(i).reshape(75, 75, 1) for i in X1])
X_R2 = np.array([np.array(i).reshape(75, 75, 1) for i in X2])
X_R1.shape
```

```python
Y = to_categorical(Y)

Y.shape
```

```python
image_height = 75
image_width = 75

train_samples = 1300
validation_samples = 304

epochs = 20
batch_size = 32
number_of_classes = 2

X_train_1 = X_R1[:train_samples, :, : ,:]
X_train_1 = abs(X_train_1/(np.max(X_train_1) - np.min(X_train_1)))
X_train_2 = X_R2[:train_samples, :, : ,:]
X_train_2 = abs(X_train_2/(np.max(X_train_2) - np.min(X_train_2)))
X_test_1 = X_R1[train_samples:, :, :, :]
X_test_1 = abs(X_test_1/(np.max(X_test_1) - np.min(X_test_1)))
X_test_2 = X_R2[train_samples:, :, :, :]
X_test_2 = abs(X_test_2/(np.max(X_test_2) - np.min(X_test_2)))

Y_train = Y[:train_samples, :]
Y_test = Y[train_samples:, :]

print(X_R1.shape)
print(X_train_1.shape)
print(X_test_1.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
%matplotlib inline
f, (ax1, ax2) = plt.subplots(1, 2, sharex=True)

sample_image = (X_train_1[0])
sample_image = np.reshape(sample_image, (75,75))
ax1.imshow(sample_image)
ax1.set_title(Y_train[0][1])
sample_image = (X_train_1[1])
sample_image = np.reshape(sample_image, (75,75))
ax2.imshow(sample_image)
ax2.set_title(Y_train[1][1])
```

```python
# Alternative go at multiple inputs

band_input_B1 = Input(shape=(image_width, image_height, 1))
band_input_B2 = Input(shape=(image_width, image_height, 1))

# Band 1 processing
CNN_L1_B1 = Conv2D(32, (3, 3), activation='relu')(band_input_B1)
CNN_L1_B1 = BatchNormalization()(CNN_L1_B1)

CNN_L2_B1 = Conv2D(64, (3, 3), activation='relu')(CNN_L1_B1)
CNN_L2_B1 = BatchNormalization()(CNN_L2_B1)
CNN_L2_B1 = MaxPooling2D(pool_size=(2, 2))(CNN_L2_B1)

CNN_L3_B1 = Conv2D(128, (3, 3), activation='relu')(CNN_L2_B1)
CNN_L3_B1 = BatchNormalization()(CNN_L3_B1)

CNN_L4_B1 = Conv2D(256, (3, 3), activation='relu')(CNN_L3_B1)
CNN_L4_B1 = BatchNormalization()(CNN_L4_B1)

CNN_L5_B1 = Conv2D(512, (3, 3), activation='relu')(CNN_L4_B1)
CNN_L5_B1 = BatchNormalization()(CNN_L5_B1)
CNN_L5_B1 = MaxPooling2D(pool_size=(2, 2))(CNN_L5_B1)

CNN_OUT_B1 = Flatten()(CNN_L5_B1)
CNN_OUT_B1 = BatchNormalization()(CNN_OUT_B1)
CNN_OUT_B1 = Dropout(0.5)(CNN_OUT_B1)

# Band 2 processing
CNN_L1_B2 = Conv2D(32, (3, 3), activation='relu')(band_input_B2)
CNN_L1_B2 = BatchNormalization()(CNN_L1_B2)

CNN_L2_B2 = Conv2D(64, (3, 3), activation='relu')(CNN_L1_B2)
CNN_L2_B2 = BatchNormalization()(CNN_L2_B2)
CNN_L2_B2 = MaxPooling2D(pool_size=(2, 2))(CNN_L2_B2)

CNN_L3_B2 = Conv2D(128, (3, 3), activation='relu')(CNN_L2_B2)
CNN_L3_B2 = BatchNormalization()(CNN_L3_B2)

CNN_L4_B2 = Conv2D(256, (3, 3), activation='relu')(CNN_L3_B2)
CNN_L4_B2 = BatchNormalization()(CNN_L4_B2)

CNN_L5_B2 = Conv2D(512, (3, 3), activation='relu')(CNN_L4_B2)
CNN_L5_B2 = BatchNormalization()(CNN_L5_B2)
CNN_L5_B2 = MaxPooling2D(pool_size=(2, 2))(CNN_L5_B2)

CNN_OUT_B2 = Flatten()(CNN_L5_B2)
CNN_OUT_B2 = BatchNormalization()(CNN_OUT_B2)
CNN_OUT_B2 = Dropout(0.5)(CNN_OUT_B2)

MERGE_LAYER = concatenate([CNN_OUT_B1, CNN_OUT_B2])

# Deep layers processing
DL = Dense(256, activation='relu', use_bias=True, bias_initializer='zeros')(MERGE_LAYER)
DL = BatchNormalization()(DL)

OUTPUT = Dense(number_of_classes, activation='softmax')(DL)

model = Model(inputs = [band_input_B1, band_input_B2], outputs=OUTPUT)

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
# model.summary()
```

```python
model.fit([X_train_1, X_train_2], Y_train,
              batch_size=batch_size, validation_data=([X_test_1, X_test_2], Y_test),
              epochs=epochs,
                shuffle=True)
```

```python
Y_pred = model1.predict(X_test)
```

```python
for i in range(len(Y_pred)):
    Y_pred[i, :] = np.round(Y_pred[i, :])
```

```python
#Precision: how many 1 are true 1?
#Recall: how many 0 are true 0? 
#F1Score: balance between those two

y_comp = confusion_matrix(Y_test[:, 1], Y_pred[:, 1])
precision = y_comp[1, 1] / (y_comp[0, 1] + y_comp[1, 1])
recall = y_comp[1, 1] / (y_comp[1, 0] + y_comp[1, 1])
accuracy = accuracy_score(Y_test[:, 1], Y_pred[:, 1])
f1score = 2 * precision * recall / (precision + recall)
print("AC:%.2f PR:%.2f RE:%.2f F1:%.2f" % (accuracy, precision, recall, f1score))
```

```python
filename = ("model_%i_%i_%i_%i.h5" % (accuracy*100, precision*100, recall*100, f1score*100))
model1.save(filename)
```
